Creating a connection to the database

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
from getpass import getpass

# entering password
password = getpass() 

#include password into connection string
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila' 

# creating a connection engine to sakila database
engine = create_engine(connection_string) 

# check if connection works:
df = pd.read_sql_query('SELECT * FROM film', engine)
df.head(3)
```

#### 1. How many distinct (different) actors' last names are there?

```python
# Writing the query
query = query = '''SELECT count(DISTINCT last_name) AS number_of_actors
FROM actor;'''
# Passing the query to the engine
df = pd.read_sql_query(query, engine)
df.head(3)
```

#### 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
query = query = '''SELECT *, 
CASE
WHEN WEEKDAY(rental_date) < 5 then 'workday'
ELSE 'weekend'
END AS 'day_type'
FROM rental;'''

df = pd.read_sql_query(query, engine)
df.head(3)
```

#### 3. Get all films with ARMAGEDDON in the title.

```python
query = query = '''SELECT title
FROM film
WHERE title LIKE '%%Armageddon%%';'''

df = pd.read_sql_query(query, engine)
df.head(10)
```

#### 4. Get 10 the longest films.

```python
query = query = '''SELECT * 
FROM film
ORDER BY length DESC
LIMIT 10;'''

df = pd.read_sql_query(query, engine)
df.head(10)
```

#### 5. How many films include Behind the Scenes content?

```python
query = query = '''SELECT count(*) AS number_of_movies
FROM film
WHERE special_features LIKE '%%Behind the scenes%%';'''

df = pd.read_sql_query(query, engine)
df.head(10)
```

#### 6. Which kind of movies (rating) have a mean duration of more than two hours?

```python
query = query = '''SELECT rating, AVG(length) AS average_length
FROM film
Group BY rating
HAVING AVG(length) > 120;'''

df = pd.read_sql_query(query, engine)
df.head(3)
```

#### 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.

```python
query = query = '''SELECT title, length,
CASE
WHEN length BETWEEN 1 AND 60 THEN 'Short'
WHEN length BETWEEN 61 AND 120 THEN 'Medium'
ELSE 'Long'
END AS length_rating
FROM film
WHERE length IS NOT NULL AND length <> 0
ORDER BY length;'''

df = pd.read_sql_query(query, engine)
df.head(10)
```
